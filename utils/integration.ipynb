```python
f = lambda x: math.sin(100*x) * math.exp(-x**2) * math.cos(2*x)
```

```python
a = 0
b = 3
N = 100000
h = (b - a) / N 
```

## Trapezoid

```python
I = 0
for i in range(N):
    I += (f(i*h) + f(i*h+h)) * h/2

I
```

## Simpson

```python
I = 0
for i in range(N):
    I += (f(i*h) + 4*f((i+i+1)*h/2) + f(i*h+h)) * h/6

I
```

## 3/8

```python
I = 0
for i in range(N):
    I += (f(i*h) + 3*f((2*i+i+1)*h/3) + 3*f((i+2*i+2)*h/3) + f(i*h+h)) * h/8

I
```
