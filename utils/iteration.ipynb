```python
import math
import numpy as np
import matplotlib.pyplot as plt
```

image.png

```python
f = lambda x: x + math.log(x)
```

```python
phi1 = lambda x: math.exp(-x)
dphi1 = lambda x: -math.exp(-x)
```

```python
N = 1000
a = 0.5
b = 1
r = np.linspace(a, b, N)
plt.plot(r, [f(x) for x in r])
plt.grid()
```

image.png

```python
q = max([abs(dphi1(x)) for x in r])
q
```

image.png

```python
eps = 1e-4
eps0 = b - a
abs(math.log(eps/eps0) / math.log(q))
```
