```python
from sympy import *
```

```python
x = symbols('x')
y = Function('y')(x)
v = Function('v')(x)
```

```python
y0 = 0
y1 = 1
guess = x
```

```python
expr = y.diff(x, 2) + y.diff(x)**2 - 8*exp(y-x)+23
Eq(expr, 0)
```

```python
expr = expr.subs(y, guess+v).simplify().expand()
Eq(expr, 0)
```

```python
diffeq = expr.subs(exp(v), v).subs(v.diff(x)**2, 0)
Eq(diffeq, 0)
```

```python
classify_ode(diffeq, v)
```

```python
dsolve(diffeq, v, ics={v.subs(x, 0): 0, v.subs(x, 1): 0})
```
