```python
f = lambda x: 20*(x**3) - 4*(x**2) - 5*x + 1
df = lambda x: 20*3*(x**2) - 4*2*x - 5
ddf = lambda x: 20*3*2*x - 4*2
eps = 1e-8
```

```python
def newtonMethod(x0):
    x1 = x0
    x2 = x1 - (f(x1)/df(x1))
    i = 1
    while abs(x2 - x1) > eps:
        x1 = x2
        x2 = x2 - (f(x2)/df(x2))
        i += 1
    return x2
```

```python
x0 = -1
newtonMethod(x0)
```
