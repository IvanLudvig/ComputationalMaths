```python
import math
```

```python
x = [0, 0.125, 0.25, 0.375, 0.5, 0.625, 0.75, 0.875, 1]
f = [0.5, 0.499512, 0.496124, 0.487155, 0.470588, 0.445605, 0.412903, 0.374543, 1/3]
```

```python
N = len(x)
h = x[1] - x[0]
```

## Trapezoid

image.png

```python
def trapezoid(step = 1):
    I = f[0]
    for i in range(0, N - 1, step):
        I += 2 * f[i]
    I += f[-1]
    I *= step * h / 2
    return I
```

## Simpson

image.png

```python
def simpson(step = 1):
    I = f[0]
    for i in range(0, N - 1, step):
        if i % 2 == 0:
            I += 2 * f[i]
        else:
            I += 4 * f[i]
    I += f[-1]
    I *= step * h / 3
    return I
```

```python
trapezoid(1)
```

```python
trapezoid(2)
```

```python
simpson(1)
```
